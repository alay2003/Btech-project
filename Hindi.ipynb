```python
import re

# Function to clean Hindi text
def clean_text(text):
    # Remove extra spaces, punctuation, and special characters
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = re.sub(r'[^\w\s\u0900-\u097F]', '', text)  # Remove all non-Hindi characters
    return text.strip()

# Apply the cleaning function
dataset['context'] = dataset['context'].apply(clean_text)
dataset['question'] = dataset['question'].apply(clean_text)

# Display the cleaned context and question columns
print(dataset[['context', 'question']].head())
```

```python
import unicodedata

# Function to normalize Hindi text
def normalize_text(text):
    text = unicodedata.normalize('NFKC', text)  # Normalize Unicode characters
    return text

# Apply normalization to context and question
dataset['context'] = dataset['context'].apply(normalize_text)
dataset['question'] = dataset['question'].apply(normalize_text)

# Check if normalization is applied
print(dataset[['context', 'question']].head())
```

```python
from transformers import AutoTokenizer

# Load pre-trained tokenizer (using mBERT in this case, but you can choose a different one)
tokenizer = AutoTokenizer.from_pretrained('bert-base-multilingual-cased')

# Example tokenization on context and question
dataset['context_tokenized'] = dataset['context'].apply(lambda x: tokenizer(x, truncation=True, padding='max_length', max_length=512))
dataset['question_tokenized'] = dataset['question'].apply(lambda x: tokenizer(x, truncation=True, padding='max_length', max_length=64))

# Check the tokenized outputs
print(dataset[['context_tokenized', 'question_tokenized']].head())
```

```python
# Import necessary libraries
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, Trainer, TrainingArguments
from datasets import Dataset
import pandas as pd
import json

# Load new dataset for fine-tuning
dataset = pd.read_csv('/content/xquad_hi.csv')

# Inspect dataset structure
print(dataset.columns)
print(dataset.head())

# Convert 'answers' column to JSON-like format if it's a string
dataset['answers'] = dataset['answers'].apply(json.loads)

# Convert DataFrame to Hugging Face Dataset object
dataset = Dataset.from_pandas(dataset)

# Load pre-trained tokenizer and the already fine-tuned model
model_name = "/content/fine_tuned_hindi_qa_model"  # Replace with the path to the model directory
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)

# Tokenizing function
def tokenize_function(examples):
    return tokenizer(
        examples["question"],
        examples["context"],
        truncation=True,
        padding="max_length",
        max_length=512
    )

# Tokenize the dataset
tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Add start and end positions for the answers
def add_start_end_positions(examples):
    start_positions = []
    end_positions = []
    for i in range(len(examples['answers'])):
        answer = examples['answers'][i]['text'][0]
        start_pos = examples['answers'][i]['answer_start'][0]
        end_pos = start_pos + len(answer) - 1
        start_positions.append(start_pos)
        end_positions.append(end_pos)
    examples['start_positions'] = start_positions
    examples['end_positions'] = end_positions
    return examples

# Add start and end positions
tokenized_dataset = tokenized_dataset.map(add_start_end_positions, batched=True)

# Set up training arguments for further fine-tuning
training_args = TrainingArguments(
    output_dir="./results",  # Output directory
    evaluation_strategy="epoch",  # Evaluate after each epoch
    learning_rate=2e-5,  # Learning rate for continued fine-tuning
    per_device_train_batch_size=16,  # Batch size for training
    per_device_eval_batch_size=16,  # Batch size for evaluation
    num_train_epochs=3,  # Number of epochs for additional training
    weight_decay=0.01,  # Weight decay for regularization
    logging_dir='./logs',  # Logging directory
    save_steps=500,  # Save checkpoint every 500 steps
    resume_from_checkpoint=True  # Resume training from last checkpoint if available
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    eval_dataset=tokenized_dataset  # Replace with validation set if available
)

# Continue training
trainer.train()
```

```python
# Function to get the answer from the model
def answer_question(question, context):
    # Tokenize the input question and context
    inputs = tokenizer(question, context, return_tensors="pt", truncation=True, padding=True, max_length=512)

    # Perform inference
    with torch.no_grad():
        outputs = model(**inputs)
        start_scores = outputs.start_logits
        end_scores = outputs.end_logits

    # Get the most likely start and end tokens
    start_index = torch.argmax(start_scores)
    end_index = torch.argmax(end_scores)

    # Debugging code to print tokens and validate indices
    tokens = tokenizer.convert_ids_to_tokens(inputs.input_ids[0])
    print("Tokens:", tokens)
    print("Start token:", tokens[start_index.item()])
    print("End token:", tokens[end_index.item()])

    # Check if the end token comes after the start token
    if start_index <= end_index:
        # Extract and decode the answer
        answer_tokens = inputs.input_ids[0][start_index:end_index + 1]
        answer = tokenizer.decode(answer_tokens, skip_special_tokens=True)
        print("Extracted answer:", answer)
    else:
        answer = ""

    return answer

# Example context and question
context = "पैंथर्स की डिफ़ेन्स ने लीग में केवल 308 अंक दिए..."
question = "पैंथर्स डिफ़ेंस ने कितने अंक दिए?"

# Get the answer from the model
answer = answer_question(question, context)
print(f"Answer: {answer}")
```
